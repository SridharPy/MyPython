```python
import pandas
df_del = pandas.read_csv("supermarkets.csv")
df_del = df_del.set_index("Address")
df_del
```

```python
#We use drop() function
#0 for rows, 1 for column
#Dropping Rows and Columns from a datframe usning names, drop function doesn't perfrom an inplace operation, we again use dataframe name as variable to perfrom actual delete
df_del.drop("332 Hill St",0)
```

```python
#To perfrom and inplace operation on delete rows
df_del = df_del.drop("332 Hill St",0)
df_del
```

```python
#Dropping state column, 1 is for column deletion
df_del.drop("State",1)
```

```python
df_del
```

```python
#to make an inplace operation for column deletion
df_del.drop("State",1)
```

```python
# To delete a column using index postion instead of name
df_del.drop(df_del.columns[2],1)
```

```python
#To drop multiple column

df_del.drop(df_del.columns[0:2],1)
```

```python
#To Drop rows using index position

df_del.drop(df_del.index[1:3],0)
```
