```python
#geopy python pacjakge has the coordinates for the addresses supplied to it
#geocoders module requires an active internet connection
#Nominatim is a type in geocoders module and geocode method in it provides the coordinates of the given address
#It accepts Road Name, City Name, State ZIP, Country  
#And returns complete address with co-ordinates 


import geopy
from geopy.geocoders import Nominatim

nom = Nominatim(scheme = "http")
cor=nom.geocode("China Town, Kolkata")
lat = cor.latitude
lon = cor.longitude
print("Kolkata Co-ordinate details :",cor,"\nLatitude of Kolkata :",lat,"\nLongitude of Kolkata :",lon)
```

```python
tgp=type(geopy)
tgr=type(geopy.geocoders)
tn=type(Nominatim)
tg =type(Nominatim.geocode)

print("Type of all GeoPy Objects:  " + "\ngeopy %s: " %tgp, "\ngeocoders %s: " %tgr, "\nNominatim %s: " %tn , "\ngeocode %s: " %tg, "\nlatitude %s" %type(cor.latitude)) 
```

```python
#Using Geopy in Pandas

import pandas

df1= pandas.read_csv("supermarkets.csv")
df1
```

```python
#geocodes function takes in parameter in the form of "Road Name, City Name, State ZIP, Country"
#Making Address column in that format

df1["Address"] = df1["Address"] +", "+df1["City"]+", "+df1["State"]+", "+df1["Country"]
df1
```

```python
#Now we need to supply each of these Address values into geocode function in Nominatim
#But Pandas don't need iteration through for
#Automatic iteration or loading of rows of a column or series happens with "apply" function dfName[ColumnName].apply(functionName)
#We just need to pass in the function as parameter in the apply to have the series / row iterate through
#Lets add a new column Location in the df1 data frame showing location details of each Address using geocode function of geopy
```

```python
df1["Cordinates"]=df1["Address"].apply(nom.geocode)#no need to use () after geocode function, "apply" automatically takes it
df1
```

```python
#But the latitude and logitude values are hidden due to long string value of Location field rows

df1.Cordinates
```

```python
#We can see the complete value of Location of any one location object
df1.Cordinates[0]
```

```python
#To see Latitude and Longitude of any one location object in the row
print(df1.Cordinates[0].latitude, df1.Cordinates[0].longitude)
```

```python
CoL = type(df1.Cordinates[0])
CoS = type(df1.Cordinates)

print("Type of first object is %s" %CoL, "\nType of second object is %s"%CoS)
```

```python
#Now we want to add two more columns Latitude and Longitude for each row
#We can't use df1["Latitude"]=df1["Cordinates"].Latitude as that will be series objects and latitude will not be applicable
#For series we need to use "apply" method
#As we are using Cordinates values here which is a Location object and we need to access latitude value from that object
#We use a special function lambda and supply temp variable x to store the vlaue of each Location
#Since the second value in df1 is None Type we use a if condition to avoid error as NoneType.latitude will throw error

df1["Latitude"] = df1["Cordinates"].apply(lambda x:x.latitude if x!=None else None)
df1["Longitude"]= df1["Cordinates"].apply(lambda x:x.longitude if x!=None else None)
df1
```
